```python
import json

def lerDataset (fnome):
    f=open (fnome,encoding="UTF-8")
    bd=[]
    dicjson=json.load(f)
    for linha in dicjson:
        bd.append(linha)
    i=0
    for entrada in bd:
        entrada["id"]=str(i)
        i+=1
    return (bd)

cinema=[]
cinema=lerDataset("cinemaATP.json")
#print (cinema)

#gravar informacao do ficheiro
def guardarDataset(bd,fnome):
    f=open (fnome,"w",encoding="UTF-8")
    json.dump (bd,f,ensure_ascii=False,indent=4)
guardarDataset(cinema,"BDcinema")

#carregar informacao do ficheiro
def carregarDataset (fnome):
    bd=[]
    f=open (fnome,encoding="UTF-8")
    bd=json.load (f)
    return (bd)
carregarDataset ("BDcinema")

#Distribuições
#numero de filmes da bd

def numFilmes(bd):
    a=0
    i=0
    for i in range (len(bd)-1):
        a+=1
    return a

BDcinema=carregarDataset("BDcinema")
#print(numFilmes(BDcinema) )


#Distribuição por género
def distribPorGenero(bd):
    distribuicao = {}
    for filme in bd:
        i=0
        for i in range(len(filme["genres"])-1):
            if filme["genres"][i] in distribuicao.keys():
                distribuicao [filme["genres"][i] ]+=1
            else:
                distribuicao [filme["genres"][i]]= 1
    return distribuicao
BDcinema=carregarDataset("BDcinema")
distribPorGenero(BDcinema)  

#Gráfico distribuicao por genero
import matplotlib.pyplot as plt
def plotDistribPorGenero(bd):
    fig=plt.figure(figsize=(7,7))
    ax = fig.add_axes([0,0,1,1])
    x=[]
    y=[]
    distribuicao=distribPorGenero(bd)
    for genero in distribuicao:
        x.append (genero)
        y.append (distribuicao[genero])
    plt.barh(x,y,color="Green")  #permite fazer um grafico horizontal
    plt.title("Distribuição por Géneros",color="Green",size=16)
    plt.ylabel("Géneros",color="Green",size=13)
    plt.xlabel("Número de filmes",color="Green",size=13)
    plt.show()
plotDistribPorGenero(BDcinema)

#Distribuição por ator-top10(falta)
def distribPorAtor(bd):
    distribuicao = {}
    for filme in bd:
        i=0
        for i in range(len(filme["cast"])-1):
            if filme["cast"][i] in distribuicao.keys():
                distribuicao [filme["cast"][i] ]+=1
            else:
                distribuicao [filme["cast"][i]]= 1
    return distribuicao
BDcinema=carregarDataset("BDcinema")
distribPorAtor(BDcinema) 

def chaveOrd(dis):
    return dis[1]

def top100 (dis):
    ordenado=dis.sort(dis.items(),key=chaveOrd,reverse=True)
    for i in range (100):
        print (ordenado[i])
#top100(distribPorAtor(BDcinema))
    
        
#nota, o inserir tem de confirmar se o filme ja esta incluido ou nao na lista!       
    
```
