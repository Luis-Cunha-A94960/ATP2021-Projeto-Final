```python
import json

def lerDataset (fnome):
    f=open (fnome,encoding="UTF-8")
    bd=[]
    dicjson=json.load(f)
    for linha in dicjson:
        bd.append(linha)
    i=0
    for entrada in bd:
        entrada["id"]=str(i)
        i+=1
    return (bd)

cinema=[]
cinema=lerDataset("cinemaATP.json")
#print (cinema)

#gravar informacao do ficheiro
def guardarDataset(bd,fnome):
    f=open (fnome,"w",encoding="UTF-8")
    json.dump (bd,f,ensure_ascii=False,indent=4)
guardarDataset(cinema,"BDcinema")

#carregar informacao do ficheiro
def carregarDataset (fnome):
    bd=[]
    f=open (fnome,encoding="UTF-8")
    bd=json.load (f)
    return (bd)
carregarDataset ("BDcinema")

#Distribuições
#numero de filmes da bd

def numFilmes(bd):
    a=0
    i=0
    for i in range (len(bd)-1):
        a+=1
    return a

BDcinema=carregarDataset("BDcinema")
#print(numFilmes(BDcinema) )


#Distribuição por género
def distribPorGenero(bd):
    distribuicao = {}
    for filme in bd:
        i=0
        for i in range(len(filme["genres"])-1):
            if filme["genres"][i] in distribuicao.keys():
                distribuicao [filme["genres"][i] ]+=1
            else:
                distribuicao [filme["genres"][i]]= 1
    return distribuicao
BDcinema=carregarDataset("BDcinema")
#distribPorGenero(BDcinema)  

#Gráfico distribuicao por genero
import matplotlib.pyplot as plt
def plotDistribPorGenero(bd):
    fig=plt.figure(figsize=(7,7))
    ax = fig.add_axes([0,0,1,1])
    x=[]
    y=[]
    distribuicao=distribPorGenero(bd)
    for genero in distribuicao:
        x.append (genero)
        y.append (distribuicao[genero])
    plt.barh(x,y,color="Green")  #permite fazer um grafico horizontal
    plt.title("Distribuição por Géneros",color="Green",size=16)
    plt.ylabel("Géneros",color="Green",size=13)
    plt.xlabel("Número de filmes",color="Green",size=13)
    plt.show()
#plotDistribPorGenero(BDcinema)

#Distribuição por ator e top100
def distribPorAtor(bd):
    distribuicao = {}
    for filme in bd:
        i=0
        for i in range(len(filme["cast"])-1):
            if filme["cast"][i] in distribuicao.keys():
                distribuicao [filme["cast"][i] ]+=1
            else:
                distribuicao [filme["cast"][i]]= 1
    return distribuicao
BDcinema=carregarDataset("BDcinema")
#print(distribPorAtor(BDcinema))

def top50 (dis):
    dis.pop ("and")
    dis.pop ("the")
    dis.pop ("(voice)")
    ordenado=sorted(dis.items(),key=lambda x:x[1],reverse=True)
    top50=[]
    for i in range (50):
        top50.append(ordenado[i]) 
    return (top50)
#print(top50(distribPorAtor(BDcinema)))
    
def plotDistribPorAtor(bd):
    fig=plt.figure(figsize=(8,8))
    ax = fig.add_axes([0,0,1,1])
    x=[]
    y=[]
    distribuicao= top50(distribPorAtor(bd))
    for tuplo in distribuicao:
        lista=list(tuplo)
        x.append (lista[0])
        y.append (lista[1])
    plt.barh(x,y,color="Green")  #permite fazer um grafico horizontal
    plt.title("Distribuição por Ator",color="Green",size=16)
    plt.ylabel("Ator",color="Green",size=13)
    plt.xlabel("Número de filmes",color="Green",size=13)
    plt.show()       
      
#plotDistribPorAtor(BDcinema)  

#inserir um novo filme
#nota, o inserir tem de confirmar se o filme ja esta incluido ou nao na lista! 
import re
def inserir(bd, titulo, ano, elenco, genero):
    procura=False
    if ", " in elenco:
        elenco=elenco.split(", ")

    if ", " in genero:
        genero=genero.split(", ")
    if (titulo=="" or elenco=="" or genero=="" or not re.match("^[0-9]*$", ano)):
        mensagem = "Erro na introdução do novo filme: algum dos campos foi mal preenchido!"
    for filme in bd:
        if (ano == filme["year"] or titulo == filme["title"] or elenco == filme["cast"] or genero == filme["genres"]):
            mensagem = "Erro na introdução do novo filme: o filme já existe!"
            procura=True
    if procura == False:
        mensagem = "Sucesso na inserção do novo filme!"
        bd.append({"title":titulo, "year":int(ano), "cast":elenco, "genres":genero})
        guardarDataset(bd,"BDcinema")   
    return mensagem
print(inserir(BDcinema, "unga bunga", "2018", "Big Poppa", "Comedy"))


def consultar(bd, idf, nomef):
    Resultpesquisa=0
    i=0
    if (idf == "" and nomef == "")or(idf != "" and nomef != ""): #o utilizador não deverá preencher ambos os campos nem deixar ambos por preencher
        Resultpesquisa="Erro na inserção dos dados, consulte, por favor, as informações."
    else:
        while (i <len(bd) and Resultpesquisa==0):
            if nomef == bd[i]["title"] and idf == "":
                Resultpesquisa=bd[i]
            elif idf == bd[i]["id"] and nomef == "":
                Resultpesquisa=bd[i]
            i+=1
        if Resultpesquisa==0:
            Resultpesquisa="Não existe nenhum filme com esse nome ou ID! Para adicionar um novo filme prima: Inserir."
    return Resultpesquisa
#print(consultar(BDcinema,"","102 Dalmatians"))
    

#Informações sobre o programa
def info():
    info = """
    Informações:
    
    -Consultar: deixa ao utilizador pesquisar por um filme através do seu Nome ou ID (não os dois ao mesmo tempo);
    -Carregar: carrega o ficheiro contendo o catálogo de filmes;
    -Guardar: grava o atual catálogo de filmes no ficheiro \"BDcinema\";
    -Inserir: insere um novo filme no catálogo de filmes, deverá usar um \", \" (vírgula + espaço) para separar os elementos do elenco e os géneros;
    -Listar: organiza (por ordem alfabética, género ou ator) e mostra ao utilizador os dados organizados;
    -Dados estatísticos: mostra dados estatísticos, como o número de filmes com um dado género ou o número de filmes em que cada ator participou, através de gráficos;
    -Sair: sai da aplicação.
    """
    return info
      
```
